```python
def func(r):
    if r.salary.strip() == "low":
        if r.satisfaction_level > .9:
            return "medium-new"
        else:
            return r.salary
    return r.salary

data['salary'] = data.apply(func, axis=1)
```

```python
data[data.salary == "medium-new"]
```

```python
data['salary'] = data.salary.map(lambda s: "medium" if s == "medium-new" else s)
```
